1. **Identify the numerical columns**: Select columns that contain numerical data for normalization.
2. **Calculate the mean and standard deviation**: Compute the mean and standard deviation for each numerical column.
3. **Apply normalization**: Subtract the mean from each value and then divide by the standard deviation.

Let's perform these steps using Python.

### Explanation of the Steps

1. **Identify the Numerical Columns**:
   We selected columns containing numerical data (integers and floats) for normalization.

2. **Calculate the Mean and Standard Deviation**:
   We computed the mean and standard deviation for each numerical column. These statistics are essential for the normalization process.

3. **Apply Normalization**:
   For each value in the numerical columns, we subtracted the column mean and then divided by the column standard deviation. This process transformed the data to have a mean of 0 and a standard deviation of 1, effectively normalizing it.

### Code Explanation

- `data.select_dtypes(include=['float64', 'int64']).columns`: This line identifies the numerical columns in the dataset.
- `data[numerical_columns].mean()`: Computes the mean of each numerical column.
- `data[numerical_columns].std()`: Computes the standard deviation of each numerical column.
- `(data[numerical_columns] - means) / std_devs`: Applies the normalization formula to each value in the numerical columns.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Load the dataset
df = pd.read_csv('normalized.csv')

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())

# Data Preprocessing
# Checking for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Dropping rows with missing values (if any)
df.dropna(inplace=True)

# Assuming the target column is named 'target'
# Replace 'target' with the actual name of the target column if different
target_column = 's2'  # Update this with the actual target column name
X = df.drop(target_column, axis=1)
y = df[target_column]

# Ensure the target is binary
# Convert the target to binary if it is continuous
# Here we use a threshold of 0.5 for example, you can adjust based on your needs
y_binary = (y > 0.5).astype(int)

# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_binary, test_size=0.2, random_state=42)

# Build the Deep Learning model
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=5, batch_size=10, validation_data=(X_test, y_test))

# Make predictions on the test set
y_pred = (model.predict(X_test) > 0.5).astype("int32")

# Print predictions
print("\nPredictions on the test set:")
print(y_pred)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Accuracy, Precision, Recall
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print(f"\nAccuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")

# Plotting the accuracy and loss
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.title('Model Accuracy')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.title('Model Loss')

plt.show()
```
